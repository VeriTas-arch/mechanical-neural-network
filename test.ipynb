```python
import matplotlib.pyplot as plt
import numpy as np
import math

def rms_func(rms):
    """process the root mean square error"""
    # approximate the Dirac delta function
    a = 1
    gauss = math.exp(-(rms ** 2) / (a ** 2)) / (a * math.sqrt(math.pi))
    return gauss

x = np.arange(-2, 2, 0.1)
y = [rms_func(i) for i in x]

plt.plot(x, y)
plt.show()
```
